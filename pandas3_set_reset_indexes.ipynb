##Indexes - How to Set, Reset, and Use Indexes
Usado o tutorial [Python Pandas Tutorial (Part 3): Indexes - How to Set, Reset, and Use Indexes](https://youtu.be/W9XjRYFkkyw).

```python
import pandas as pd
```

```python
# Neste caso, há sobrenomes iguais. A principio, a serie de email possui valor unico para cada linha.
people = {
    "primeiro": ["Julia", "Amanda", "Roberto"],
    "ultimo": ["Marinho", "Maciel", "Maciel"],
    "email": ["julia.marinho@gmail.com", "amanda.maciel@hotmail.com", "roberto.silva@yahoo.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
# Trocar o index com integer para uma coluna a minha escolha
df.set_index('email', inplace=True)
df
```

```python
df.index
```

O index é um identificador único para cada linha.
Se o index é o label para cada linha, usar o loc é muito interessante.

```python
df.loc['roberto.silva@yahoo.com']
```

```python
# resetar o index
df.reset_index(inplace=True)
df
```
