##Filtering - Using Conditionals to Filter Rows and Columns
Usado o tutorial [Python Pandas Tutorial (Part 4): Filtering - Using Conditionals to Filter Rows and Columns](https://youtu.be/Lw2rlcxScZY).

```python
import pandas as pd
```

```python
# Neste caso, há sobrenomes iguais. A principio, a serie de email possui valor unico para cada linha.
people = {
    "primeiro": ["Julia", "Amanda", "Roberto"],
    "ultimo": ["Marinho", "Maciel", "Maciel"],
    "email": ["julia.marinho@gmail.com", "amanda.maciel@hotmail.com", "roberto.silva@yahoo.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
# filter = (df['ultimo'] == 'Maciel') & (df['primeiro'] == 'Amanda')
filter = (df['ultimo'] == 'Bianca') | (df['primeiro'] == 'Amanda')
```

```python
# O filtro "marca" as linhas com True ou False.
filter
```

```python
# O DataFrame imprime apenas as linhas com True
df[filter]
```

```python
# df.loc[filter, 'email']
df.loc[~filter, 'email'] #negação do filter
```
