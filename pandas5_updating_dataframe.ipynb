##Updating Rows and Columns
Usado o tutorial [Python Pandas Tutorial (Part 5): Updating Rows and Columns - Modifying Data Within DataFrames](https://youtu.be/DCDe29sIKcE).

```python
import pandas as pd
```

```python
# Neste caso, há sobrenomes iguais. A principio, a serie de email possui valor unico para cada linha.
people = {
    "primeiro": ["Julia", "Amanda", "Roberto"],
    "ultimo": ["Marinho", "Maciel", "Maciel"],
    "email": ["julia.marinho@gmail.com", "amanda.maciel@hotmail.com", "roberto.silva@yahoo.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df.columns
```

```python
df.columns = ['primeiro_nome', 'ultimo_nome', 'email']
df
```

```python
df.columns = [x.lower() for x in df.columns]
df
```

```python
df.columns = df.columns.str.replace(' ','_')
df
```

```python
df.rename(columns={'ultimo_nome':'sobrenome'}, inplace=True)
df
```

```python
df.loc[2,['sobrenome','email']] = ['Oliveira','roberto.oliveira@yahoo.com']
#df.at[2,['sobrenome','email']] = ['Oliveira','roberto.oliveira@yahoo.com'] #2a opção para atualização de dados
df
```

```python
filt = (df['email']=='roberto.oliveira@yahoo.com')
df.loc[filt,'sobrenome']='Maciel'
df
```

```python
df['sobrenome'] = df['sobrenome'].str.lower()
df
```

```python
df['email'].apply(len)
```

```python
def update_email(email):
  return email.upper()
```

```python
df['email'] = df['email'].apply(update_email)
df
```

```python
df['email'] = df['email'].apply(lambda x: x.lower())
df
```

```python
#df.apply(len) #padrão é axis='rows'
df.apply(len, axis='columns')
```

```python
df.apply(pd.Series.min)
# df.apply(lambda x: x.min()) #funciona igual a linha de cima
```

```python
df.applymap(len) #applymap aplica a função em cada elemento do dataframe individualmente
```

```python
df.applymap(str.lower)
```

```python
df['primeiro_nome'] = df['primeiro_nome'].replace({'Julia':'Milena'})
df
```
